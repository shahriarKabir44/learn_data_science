```python
ocean_proximity_dict={
    '<1H OCEAN':1,
    'INLAND':2,
    'NEAR OCEAN':3,
    'NEAR BAY':4,
    'ISLAND':5
}
```

```python
df['ocean_proximity']=df['ocean_proximity'].map(lambda place: ocean_proximity_dict[place])
```

```python
df.head()
```

```python
df.plot.scatter(x='latitude',y='median_income',c='ocean_proximity',s=1)
```
