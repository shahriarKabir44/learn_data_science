```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv('./car_crashes.csv')
```

```python
df.sample(10)
```

```python
tips_df=sns.load_dataset('tips')
```

```python
tips_df.sample(5)
```

## histogram

```python
sns.displot(df['speeding'],kde=False )
```

```python
sns.displot(data=tips_df['tip'])
```

## joint plot 
### scatterplot

```python
sns.jointplot(x="speeding",y="alcohol",data=df,kind='reg')
```

```python
df.plot.scatter(x="speeding",y="alcohol")
plt.show()
```

# pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(tips_df,hue='sex')
```

## barplot

```python
df.columns
```

```python

plt.figure(figsize=(10,10))
sns.barplot(x="abbrev",y="not_distracted",data=df )
plt.xticks(rotation=90)
plt.show()
```

```python

plt.figure(figsize=(10,10))
sns.barplot(x="sex",y="tip",data=tips_df,hue="smoker" )
plt.xticks(rotation=90)
plt.show()
```

## countplot

```python
sns.countplot(x='sex',data=tips_df)
```

## boxplot

```python
sns.boxplot(x="day",y="tip",data=tips_df,hue='sex')
```

```python
sns.boxplot(x="day",y="total_bill",data=tips_df,hue='sex')
```

## violon plot

```python
sns.violinplot(x="day",y="total_bill",data=tips_df,hue='sex')
```

## stripplot

```python
sns.stripplot(x="sex",y="total_bill",data=tips_df,hue='day')
```

# heatmaps

```python
crash_df_corr=df.corr()
crash_df_corr
```

```python
sns.heatmap(crash_df_corr,cmap='Blues')
```

## facet grid

```python
tips_fg=sns.FacetGrid(tips_df,col='time',row='smoker')
tips_fg.map(plt.hist,'total_bill' )
```
